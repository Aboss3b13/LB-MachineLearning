```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = r'C:\School\BBB\Herr Colic\M259\LB\realtor-data.zip.csv'
df = pd.read_csv(file_path)

# Remove rows where 'price' or 'bed' is missing or NaN
df = df.dropna(subset=['price', 'bed'])

# Remove rows where 'price' is zero or negative (ensure only valid prices)
df = df[df['price'] > 0]

# Remove outliers for 'price' (values that are beyond 99th percentile or below 1st percentile)
q1 = df['price'].quantile(0.01)
q3 = df['price'].quantile(0.99)
iqr = q3 - q1
df = df[(df['price'] >= q1 - 1.5 * iqr) & (df['price'] <= q3 + 1.5 * iqr)]

# Plot the histogram for the actual 'price' values (no transformations)
plt.figure(figsize=(10, 6))
plt.hist(df['price'], bins=50, color='blue', edgecolor='black')
plt.title('Distribution of House Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

# Scatter plot for 'price' vs. 'bed' (number of bedrooms) after cleaning
plt.figure(figsize=(10, 6))
plt.scatter(df['bed'], df['price'], alpha=0.5, color='green')
plt.title('Price vs. Number of Bedrooms')
plt.xlabel('Number of Bedrooms')
plt.ylabel('Price')
plt.grid(True)
plt.show()

```
